## Fishing Dataset

### Read Data

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/arteagac/xlogit/master/examples/data/fishing_long.csv")

varnames = ['price','catch']
X = df[varnames].values
y = df['choice'].values
transvars = ['price', 'catch']
```

### Fit the model

```python
# from xlogit import MixedLogit

model = MixedLogit()
model.fit(X, y, alt=['beach', 'boat', 'charter', 'pier'],   # bcvars=bcvars,
          # isvars=['income'],  # , asvars=['price'],
          # asvars=asvars,
          transvars=transvars, 
          transformation="boxcox",
          randvars={'price': 'n', 'catch': 'n'},
          varnames=varnames,  # init_coeff=[0],#init_coeff=np.zeros(X.shape[1])
          # fit_intercept=True
          )
model.summary()
```
